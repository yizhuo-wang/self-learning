- Model fitting. Maximum Likelihood Estimation (MLE) for Gaussian random fields using likfit.

```r
 ( parana.ml1 <- likfit(parana, ini.cov.pars = c(4500, 50) , cov.model = 'matern') ) # 4500 (sigmasq: variance)  50 (phi: range)
```

- ML Summary

```r
summary(parana.ml1)
```

- Model prediction
There are two functions for spatial prediction in geoR,

1.   krige.conv (conventional kriging) performs “plug-in” predictions, i.e. treating estimated 
parameters as if they were the true values.
2.   krige.bayes performs Bayesian inference and prediction.






```r
# Choose prediction locations
parana.gr <- pred_grid(parana$borders, by = 15)
points(parana, pch = 16)
points(parana.gr, pch = 19, col = 'gray', cex = 0.25)
parana.gr0 <- locations.inside(parana.gr, parana$borders)
points(parana.gr0, pch = 19, col = 'blue', cex = 0.25)
```

```r
KC <- krige.control(obj.m = parana.ml1, trend.d = "1st", trend.l = "1st")
OC <- output.control(simulations = TRUE, n.pred = 1000, quantile = c(0.1, 0.25, 0.5, 0.75, 
    0.9), threshold = 350)
parana.kc <- krige.conv(parana, loc = parana.gr, krige = KC, output = OC)
names(parana.kc)
```

```r
dim(parana.kc$simulations)
```

```r
options(repr.plot.width = 20, repr.plot.height = 16)
color = colorRampPalette(c("light green", "yellow", "orange", "red"))(100)
color2 = colorRampPalette(c("light blue", "red"))(100)


### Plot prediction mean
par(mar=c(5,5,5,13), mfrow = c(2,2), pty = 'm')
plot(parana$borders, type = 'l', main = 'Prediction Mean', cex.main = 2)
points(parana$coords, col = color[as.numeric(cut(c(parana$data,180,420),breaks = 100))], pch = 15, cex = 2)
points(parana.gr0, col = color[as.numeric(cut(c(parana.kc$predict,180,420),breaks = 100))], pch = 19, cex = 1)
par(mar=c(0,0,0,0))
fields::image.plot(legend.only=T, zlim=c(180,420), col = color)

### Plot prediction variance
par(mar=c(5,5,5,13))
plot(parana$borders, type = 'l', main = 'Prediction Variance', cex.main = 2)
points(parana$coords, col = 'black', pch = 15, cex = 2)
points(parana.gr0, col = color2[as.numeric(cut(c(parana.kc$krige.var),breaks = 100))], pch = 19, cex = 1)
par(mar=c(0,0,0,0))
fields::image.plot(legend.only=T, zlim=c(parana.kc$krige.var), col = color2)


### Plot simulated prediction data
par(mar=c(5,5,5,13))
plot(parana$borders, type = 'l', main = 'Simulated Prediction', cex.main = 2)
points(parana$coords, col = color[as.numeric(cut(c(parana$data,180,420),breaks = 100))], pch = 15, cex = 2)
points(parana.gr0, col = color[as.numeric(cut(c(parana.kc$simulations[,1],180,420),breaks = 100))], pch = 19, cex = 1)
par(mar=c(0,0,0,0))
fields::image.plot(legend.only=T, zlim=c(180,420), col = color)

### Plot another simulated prediction data
par(mar=c(5,5,5,13))
plot(parana$borders, type = 'l', main = 'Another Simulated Prediction', cex.main = 2)
points(parana$coords, col = color[as.numeric(cut(c(parana$data,180,420),breaks = 100))], pch = 15, cex = 2)
points(parana.gr0, col = color[as.numeric(cut(c(parana.kc$simulations[,2],180,420),breaks = 100))], pch = 19, cex = 1)
par(mar=c(0,0,0,0))
fields::image.plot(legend.only=T, zlim=c(180,420), col = color)
```

# Example #2

```r
library(geoR)
sigma_sq = 1
beta = 0.1
nu = 0.5  
sims = grf(n = 1600, grid = "reg", cov.pars = c(sigma_sq, beta), 	kappa = nu, RF = FALSE) # Simulation of Gaussian Random Fields
data = list(x = sims$coords[, 1], y = sims$coords[, 2], z = sims$data)
```

```r
options(repr.plot.width = 7, repr.plot.height = 5)

color = colorRampPalette(c("light green", "yellow", "orange", "red"))(100)
par(mar=c(5,5,0,8))
plot(data$x, data$y, col = color[as.numeric(cut(data$z,breaks = 100))], pch = 15, cex = 1.4)
fields::image.plot(legend.only=T, zlim=range(data$z), col = color, add = TRUE)
```

```r
start_time <- Sys.time()
fit_obj = likfit(coords = cbind(data$x,data$y),
                 data = data$z,
                 trend = "cte",
                 ini.cov.pars = c(0.001,0.001),
                 fix.nugget = TRUE,
                 nugget = 0,fix.kappa = FALSE,
                 kappa = 0.001,
                 cov.model = "matern",
                 lik.method = "ML",
                 limits = pars.limits(sigmasq = c(0.001,5),phi = c(0.001,5),
                                      kappa = c(0.001,5)),method = "Nelder-Mead",control = list(abstol = 1e-4,
                                                                                                maxit = 2000))
end_time <- Sys.time()
end_time - start_time
print(fit_obj)
```

# ExaGeoStatR
ExaGeoStatR is an R-Wrapper for [ExaGeoStat framework]((https://github.com/ecrc/exageostat), a parallel high performance unified software for geostatistics on manycore systems.

Examples from https://github.com/ecrc/exageostatR


 - Example #1  (Exact)

```r
#sysctl -n hw.ncpu
library("exageostatr")                                        #Load ExaGeoStatR lib.
seed          = 0                                             #Initial seed to generate XY locs.
sigma_sq      = 1                                             #Initial variance.
beta          = 0.1                                           #Initial range.
nu            = 0.5                                           #Initial smoothness.
dmetric      = "euclidean"                                    #"euclidean", or "great_circle".
n             = 400                                          #n*n locations grid.
exageostat_init(hardware = list (ncores=2, ngpus=0, 
                                 ts=160, pgrid=1, qgrid=1))		#Initiate exageostat instance.
data = simulate_data_exact(sigma_sq, beta, nu, dmetric, n, seed) 				#Generate Z observation vector.
result        = exact_mle(data, 
                 dmetric,
                  optimization = list(clb = c(0.001, 0.001, 0.001),
                                  cub = c(5, 5,5 ),
                                   tol = 1e-3, 
                                max_iters = 3))                #Estimate MLE parameters (Exact).
exageostat_finalize()					      #Finalize exageostat instance.
```

```r
options(repr.plot.width = 7, repr.plot.height = 5)

color = colorRampPalette(c("light green", "yellow", "orange", "red"))(100)
par(mar=c(5,5,0,8))
plot(data$x, data$y, col = color[as.numeric(cut(data$z,breaks = 100))], pch = 19, cex = 1.4)
fields::image.plot(legend.only=T, zlim=range(data$z), col = color, add = TRUE)
```

```r
print(result)
```

- Example #2  (Tile Low-Rank (TLR))

```r
library("exageostatr")                                        #Load ExaGeoStatR lib.
seed            = 0                                           #Initial seed to generate XY locs.
sigma_sq        = 1                                           #Initial variance.
beta            = 0.1                                        #Initial range.
nu              = 0.5                                         #Initial smoothness.
dmetric         = "euclidean"                                 #"euclidean", or "great_circle".
n               = 400                                         #n*n locations grid.
tlr_acc         = 7                                           #TLR accuracy 10^-(acc).
tlr_maxrank     = 300                                         #TLR Max Rank.

exageostat_init(hardware = list (ncores=2, ngpus=0, ts=320, lts=400,  pgrid=1, qgrid=1))			      #Initiate exageostat instance.

data = simulate_data_exact(sigma_sq, beta, nu, dmetric, n, seed) 					      #Generate Z observation vector.

result = tlr_mle(data, tlr_acc, tlr_maxrank, 
                 dmetric, optimization = list(clb = c(0.001, 0.001, 0.001), 
                                              cub = c(5, 5,5 ), tol = 1e-4, max_iters = 20))				      #Estimate MLE parameters (TLR).

exageostat_finalize() 				   	      #Finalize exageostat instance.
```

```r
options(repr.plot.width = 7, repr.plot.height = 5)

color = colorRampPalette(c("light green", "yellow", "orange", "red"))(100)
par(mar=c(5,5,0,8))
plot(data$x, data$y, col = color[as.numeric(cut(data$z,breaks = 100))], pch = 19, cex = 1.4)
fields::image.plot(legend.only=T, zlim=range(data$z), col = color, add = TRUE)
```

```r
print(result)
```

 - Example #3  (Diagonal Super-Tile (DST))

```r
library("exageostatr")                                       #Load ExaGeoStatR lib.
seed            = 0                                          #Initial seed to generate XY locs.
sigma_sq        = 1                                          #Initial variance.
beta            = 0.1                                        #Initial range.
nu              = 0.5                                        #Initial smoothness.
dmetric         = "euclidean"                                #"euclidean", or "great_circle".
n               = 400                                       #n*n locations grid.
dst_band         = 3                                         #Number of diagonal double tiles.

exageostat_init(hardware = list (ncores=2, ngpus=0, ts=320, lts=0, pgrid=1, qgrid=1))			      #Initiate exageostat instance.

data = simulate_data_exact(sigma_sq, beta, nu,
                                 dmetric, n, seed) 					      #Generate Z observation vector.

result = dst_mle(data, dst_band, dmetric, 
                        optimization = list(clb = c(0.001, 0.001, 0.001), 
                                            cub = c(5, 5, 5), 
                                            tol = 1e-4, max_iters = 20))				      #Estimate MLE parameters (DST).
exageostat_finalize()					      #Finalize exageostat instance.
```

```r
options(repr.plot.width = 7, repr.plot.height = 5)

color = colorRampPalette(c("light green", "yellow", "orange", "red"))(100)
par(mar=c(5,5,0,8))
plot(data$x, data$y, col = color[as.numeric(cut(data$z,breaks = 100))], pch = 19, cex = 1.4)
fields::image.plot(legend.only=T, zlim=range(data$z), col = color, add = TRUE)
```

```r
print(result)
```
